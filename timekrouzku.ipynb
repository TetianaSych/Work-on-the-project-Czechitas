```python
import pandas as pd

file_path = 'Time_krouzku.xlsx'


df = pd.read_excel('Time_krouzku.xlsx', engine='openpyxl')
print(df.head())
```

```python

time_column = 'original_time'

def get_start_time(time):
    if pd.isna(time):
        return 'unknown'
    return int(time.split('-')[0].split(':')[0])

df['start_time'] = df[time_column].apply(get_start_time)
```

```python
print(df[time_column].dtype)
```

```python

df[time_column] = df[time_column].astype(str)


def determine_time_of_day(time):
    try:
        hour = int(time.strip().split('-')[0].split(':')[0])
        if 6 <= hour < 10:
            return 'ráno'
        elif 10 <= hour < 12:
            return 'dopoledne'
        elif 12 <= hour < 17:
            return 'odpoledne'
        elif 17 <= hour <= 21:
            return 'večer'
        else:
            return 'upřesnit'
    except (ValueError, IndexError):
        return 'upřesnit'

   

df['Time of day'] = df[time_column].apply(determine_time_of_day)
```
