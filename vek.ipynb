```python
import pandas as pd
import re

file_path = 'krouzkytaboryPowerBI_Vek.xlsx'


data = pd.read_excel('krouzkytaboryPowerBI_Vek.xlsx', engine='openpyxl')
print(data.head())
```

```python
def get_categories(age_range):
    categories = set()
    age_range = str(age_range).strip()
    if age_range == '' or age_range == 'nan':
        return 'upřesnit' 
    if '+' in age_range:
        min_age = int(re.findall(r'\d+', age_range)[0])
        if min_age >= 18:
            return '7'
        for age in range(min_age, 19):
            if age >= 0 and age < 1:
                categories.add('1')
            elif age >= 1 and age < 3:
                categories.add('2')
            elif age >= 3 and age < 7:
                categories.add('3')
            elif age >= 7 and age < 10:
                categories.add('4')
            elif age >= 10 and age < 14:
                categories.add('5')
            elif age >= 14 and age < 18:
                categories.add('6')
            else:
                categories.add('7')
    else:
        numbers = re.findall(r'\d+(?:\.\d+)?', age_range)
        if len(numbers) == 1:
            min_age = float(numbers[0])
            max_age = min_age
        elif len(numbers) == 2:
            min_age, max_age = map(float, numbers)
        else:
            return 'upřesnit' 
        
        for age in range(int(min_age), int(max_age) + 1):
            if age >= 0 and age < 1:
                categories.add('1')
            elif age >= 1 and age < 3:
                categories.add('2')
            elif age >= 3 and age < 7:
                categories.add('3')
            elif age >= 7 and age < 10:
                categories.add('4')
            elif age >= 10 and age < 14:
                categories.add('5')
            elif age >= 14 and age < 18:
                categories.add('6')
            else:
                categories.add('7')
    
    return ', '.join(sorted(categories, key=int))


data['Категорії'] = data['Age_original'].apply(get_categories)


print(data.head())
```

```python
output_file_path = 'krouzkytaboryPowerBI_Vek_with_categories.xlsx'
data.to_excel(output_file_path, index=False)

print(data.head())
```
